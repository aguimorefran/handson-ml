```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

X = load_iris().data

# Plot the unlabeled data
plt.scatter(X[:, 0], X[:, 1], c='black', marker='o')
plt.grid()
plt.show()
```

```python
k = 5
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(X)
```

```python
y_pred is kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
import numpy as np

X_new = np.array([[5.1, 3.5, 1.4, 0.2],
                    [5.9, 3. , 5.1, 1.8],
                    [6.9, 3.1, 5.4, 2.1]])

kmeans.predict(X_new)
```

```python
# Plot the clustered data

for i in range(k):
    plt.scatter(X[y_pred == i, 0], X[y_pred == i, 1], label='Cluster %d' % i)
plt.legend()
plt.grid()
plt.show()

```

```python
# Compute the elbow method for k 1 to 6

k_min, k_max = 1, 6

inertia = []

for k in range(k_min, k_max + 1):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(k_min, k_max + 1), inertia)
plt.xlabel('k')
plt.ylabel('Inertia')
plt.grid()
plt.show()
```

```python
# Compute the silhouette score for k 2 to 6

from sklearn.metrics import silhouette_score
silhouette_scores = []

for k in range(2, 7):
    kmeans = KMeans(n_clusters=k)
    y_pred = kmeans.fit_predict(X)
    silhouette_scores.append(silhouette_score(X, y_pred))

plt.plot(range(2, 7), silhouette_scores)
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.grid()
plt.show()
```

```python
# Silhoutte diagram for k 2,3,4,5

from sklearn.metrics import silhouette_samples

for k in range(2, 6):
    kmeans = KMeans(n_clusters=k)
    y_pred = kmeans.fit_predict(X)
    silhouette_values = silhouette_samples(X, y_pred)
    plt.hist(silhouette_values, bins=50, alpha=0.75, label='k=%d' % k)

plt.xlabel('Silhouette Score')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```

## Clustering as preprocessing

```python
from sklearn.datasets import load_digits
X_digits, y_digits = load_digits(return_X_y=True)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits)

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state=42)
log_reg.fit(X_train, y_train)

log_reg.score(X_test, y_test)
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ("kmeans", KMeans(n_clusters=50)),
    ("log_reg", LogisticRegression())
])
     

pipeline.fit(X_train, y_train)
pipeline.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = dict(kmeans__n_clusters=range(2,100))
grid_clf = GridSearchCV(pipeline, param_grid, cv=3)
grid_clf.fit(X_train, y_train)
```

```python
grid_clf.best_params_
```

```python
grid_clf.score(X_test, y_test)
```

## Semi supervised

```python
n_labeled = 50
log_reg = LogisticRegression()
log_reg.fit(X_train[:n_labeled], y_train[:n_labeled])
log_reg.score(X_test, y_test)
```

```python
k = 50
kmeans = KMeans(n_clusters=k)
X_digits_dist = kmeans.fit_transform(X_train)
representative_digit_idx = np.argmin(X_digits_dist, axis=0)
x_representative_digits = X_train[representative_digit_idx]

# Plot the representative digits
plt.figure(figsize=(8, 2))
for index, x_representative_digit in enumerate(x_representative_digits):
    plt.subplot(k // 10, 10, index + 1)
    plt.imshow(x_representative_digit.reshape(8, 8), cmap='binary', interpolation='bilinear')
    plt.axis('off')
plt.show()
```

```python
# Manually label the representative digits
y_representative_digits = np.array([7,6,8,2,3,0,8,4,5,9,1,6,1,7,5,3,2,8,1,4,4,7,5,0,9,6,8,4,2,2,7,9,2,6,7,2,8,9,3,3, 5,1,3,5,7,1,6,8,9,0])

log_reg = LogisticRegression()
log_reg.fit(x_representative_digits, y_representative_digits)
log_reg.score(X_test, y_test)
```

```python
y_train_propagated = np.empty(len(X_train), dtype=np.int32)
for i in range(k):
    y_train_propagated[kmeans.labels_==i] = y_representative_digits[i]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train_propagated)
log_reg.score(X_test, y_test)
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)

# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_, marker='o')
plt.grid()
plt.show()
```

```python
# Now with 0.1

dbscan = DBSCAN(eps=0.1, min_samples=5)
dbscan.fit(X)

# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_, marker='o')
plt.grid()
plt.show()
```

```python
# Now with iris

X = load_iris().data
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X)

# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_, marker='o')
plt.grid()
plt.show()
```
